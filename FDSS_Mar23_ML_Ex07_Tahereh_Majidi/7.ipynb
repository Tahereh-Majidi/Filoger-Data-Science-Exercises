 
 همچنین تجمیع داد ها روی قله یا مد نمودار بیشتر است 
 یعنی پرواز های ما  یه درصد خیلی زیاد در حدود -13 تاخیر داشتند 

```python
avg_by_month = df.groupby(['month'], as_index=False).agg({'arr_delay': 'mean'}).round(2)
avg_by_month
```

```python
fig, ax1 = plt.subplots()
ax1.plot(avg_by_month['month'], avg_by_month['arr_delay']);
ax1.set_xticks(np.arange(1, 13, 1));
```

```python
sns.relplot(data=df, x='month', y='arr_delay', kind='line');
```

بیشترین میانگین تاخیر در رسیدن به مقصد در ماه هفتم و کمترین میانگین تاخیر در ماه نهم است

```python
fig, ax1 = plt.subplots()
ax1.bar(avg_by_month['month'], avg_by_month['arr_delay'])
ax1.set_xticks(np.arange(1, 13, 1));
```

همانطور که میبینیم هر چه به اخر شب نزدیک میشویم تاخیر بیشتر میشود
بیشترین تاخیر در نه شب بوده است

```python
# Create a scatter plot of departure delay versus air time
sns.set(style='whitegrid')
sns.scatterplot(x='air_time', y='dep_delay', data=df)
plt.title('Departure Delay versus Air Time')
plt.xlabel('Air Time (minutes)')
plt.ylabel('Departure Delay (minutes)')
plt.show()
```

همانطور که میبینیم هر چقدر زمان کمتری پرواز طول بکشد تعداد تاخیر بیشتر است.
ولی  اکثرا زمان تاخیر کمتر است 

```python

# Create a scatter plot of departure delay versus distance
sns.set(style='whitegrid')
sns.scatterplot(x='distance', y='dep_delay', data=df)
plt.title('Departure Delay versus distance')
plt.xlabel('distance (in miles)')
plt.ylabel('Departure Delay (minutes)')
plt.show()
```

همچنین هر چه فاصله کمتر باشد تعداد تاخیر بیشتر است 
در فاصله های بیشتر تعداد تاخیر کمتر است
